```python
# Problem 2

def lfsr_sequence(initial_state, feedback_function, length=15):
    state = initial_state[:]
    period = []
    seen_states = []

    while tuple(state) not in seen_states:
        seen_states.append(tuple(state))
        period.append(state[0])
        # Beregn neste bit basert på feedback-funksjonen
        next_bit = sum(state[i] for i in feedback_function) % 2
        state = state[1:] + [next_bit]

    return period, len(seen_states)

# Initialisering av nøklene
key_1 = [1, 0, 0, 0]
key_2 = [0, 0, 1, 1]

# Kjør LFSR for de gitte nøklene og feedback-funksjonene
feedback_a = [0, 1, 2, 3]  # z_{i+4} = z_i + z_{i+1} + z_{i+2} + z_{i+3}
feedback_b = [0, 3]        # z_{i+4} = z_i + z_{i+3}

# Simulering
for key, keyname in zip([key_1, key_2], ["1000", "0011"]):
    period_a, length_a = lfsr_sequence(key, feedback_a)
    print(f"Period for key {keyname} with LFSR a): {length_a}, sequence: {period_a}")

    period_b, length_b = lfsr_sequence(key, feedback_b)
    print(f"Period for key {keyname} with LFSR b): {length_b}, sequence: {period_b}")
```
